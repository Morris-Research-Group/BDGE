```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# very simple and not necessarily sophisticated computation time comparison


# We did time the differential gene expression analysis in our bootstrapping experiments in

# do_DGE.ipynb 

# and read out the duration in seconds.
```

```python
data = pd.DataFrame({'edgeR': [216, 238, 284, 333],
                     'DESeq2': [778, 837, 1035, 1176],
                     'bayexpress': [4, 5, 5, 6]})

data
```

```python
fig, axs = plt.subplots(1, 1, sharey=True, figsize=(5, 5), dpi=300)
axs.grid(alpha=0.4)
# scatters
axs.scatter(np.arange(1, 5),
            data.edgeR,
            c='#117733', s=50,
            label='edgeR',
            alpha=0.9, marker='x')

axs.scatter(np.arange(1, 5),
            data.DESeq2,
            c='#332288', s=50,
            label='DESeq2',
            alpha=0.9, marker='x')

axs.scatter(np.arange(1, 5),
            data.bayexpress,
            c='#882255', s=50,
            label='bayexpress',
            alpha=0.9, marker='x')

axs.legend(loc='upper left')


# Use tex in labels
axs.set_xticks(np.arange(1, 5))
axs.set_xticklabels(['3 reps', '6 reps', '12 reps', '20 reps'])

# Rotate the tick labels and set their alignment.
plt.setp(axs.get_xticklabels(), rotation=45, ha="right",
        rotation_mode="anchor")


axs.set_xlabel('')
axs.set_ylabel('seconds runtime')



plt.show()
```
